```python
from nbgrader.api import Gradebook
import pandas as pd
import macti.grader as mgrad
```

```python
# Archivo CSV con la información de los estudiantes proveniente de Moodle (Participants)
students_df = pd.read_csv("hecompa2025_courseid_4_participants.csv") 
students_df
```

```python
db_name = "sqlite:///gradebook.db"
print(f"\nBase de datos: {db_name}")
```

```python
with Gradebook(db_name) as gb:
    # Mostrar información de los estudiantes en la base de datos de nbgrader
    print(mgrad.get_students_df(gb))

    # Actualizar la info de los estudiantes desde Moodle hacia nbgrader
    mgrad.update_students(gb, students_df)

    # Mostrar la información de los estudiantes en la base de datos de nbgrader actualizada
    print(mgrad.get_students_df(gb))
```

```python
with Gradebook(db_name) as gb:
    # Mostrar la información de los assignments en la base de datos
    mgrad.show_assignments(gb)
    
    # Actualizar la lista de assignments
    mgrad.update_assignments(gb)
        
    # Mostrar la información de los assignments en la base de datos
    mgrad.show_assignments(gb)
```
