```python
A = np.array([[0.10, -1.],[0.30,-1.]] )
B = np.array([-200, 20])
C = 0
```

```python
file_answer.write('4', A, 'Checa las entradas de la matriz A')
file_answer.write('5', B, 'Checa las entradas del vector B')
file_answer.write('6', C, 'Checa el valor de C')
```

### Forma cuadrática

$$ f(\mathbf{x}) = \dfrac{1}{2} \mathbf{x}^T A \mathbf{x} - \mathbf{b}^T \mathbf{x} + c $$

Usando los valores de $A$, $B$ y $C$ calcula la forma de $f(\mathbf{x})$
